# MMF Stage 2 results summary

Stage 2 的重点不是显式恢复 transmission matrix，而是验证 CNN/ResNet 能否直接学习 speckle-to-image 的逆映射。

结果说明，在训练和测试光纤状态匹配时，static model 可以恢复输入图案；但当 kappa 改变后，重建质量下降。这也是 Stage 3 需要引入动态更新的原因。

## 本次结果

| condition | pixel accuracy | correlation |
| --- | ---: | ---: |
| matched-state test | 0.9943 | 0.9878 |
| kappa = 0.0 | 0.9964 | 0.9910 |
| kappa = 3.0 | 0.8402 | 0.7508 |

上表汇总的是前两个 notebook 直接显示的结果。这个 summary 不再读取或写入额外结果文件。

## Poster wording

> The static CNN/ResNet inverse model reconstructs input patterns under the matched fiber state, but its performance drops when kappa shifts. This suggests that the learned inverse mapping is tied to a fixed MMF transmission state, motivating dynamic adaptation in Stage 3.
